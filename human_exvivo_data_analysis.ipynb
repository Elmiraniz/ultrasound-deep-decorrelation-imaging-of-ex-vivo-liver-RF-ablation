```python
import numpy as np
import matplotlib.pyplot as plt
```

# Data analysis
## Analysis of number of tissue samples per group

```python
uncont_samples = {'Normal':3, 'Steatotic':6, 'Cirrhotic':2}
cont_samples = {'Normal':6, 'Steatotic':7, 'Cirrhotic':4}
```

```python
conditions = list(uncont_samples.keys())
uncont_sample_values = list(uncont_samples.values())
cont_sample_values = list(cont_samples.values())

font_size = 20
barWidth = 0.25
br1 = np.arange(len(conditions))
br2 = [x + barWidth for x in br1]

fig = plt.figure(figsize = (10, 5))
 
# Make the plot
plt.bar(br1, uncont_sample_values, color ='k', width = barWidth,
        edgecolor ='grey', label ='Not cont.')
plt.bar(br2, cont_sample_values, color ='gray', width = barWidth,
        edgecolor ='grey', label ='Cont.')
 
# Adding Xticks
plt.xlabel('Tissue condition', fontsize = font_size)
plt.ylabel('Number of samples', fontsize = font_size)
plt.xticks([r + barWidth/2 for r in range(len(conditions))],conditions)
plt.legend(fontsize = font_size)
plt.show()
```

## Analysis of final decorrelation values

```python
uncont_normal_finaldec = [-2.768, -2.52, -2.4]
uncont_steatotic_finaldec = [-3.264, -2.69, -2.19, -2.64, -2.918, -2.298]
uncont_cirrhotic_finaldec = [-2.15, -1.95]

barWidth = 0.25
br1 = np.arange(len(conditions))

fig = plt.figure(figsize = (10, 5))
 
plt.bar(br1, [np.mean(uncont_normal_finaldec), np.mean(uncont_steatotic_finaldec), np.mean(uncont_cirrhotic_finaldec)],
        color ='gray', width = barWidth, edgecolor ='grey', 
        yerr = [np.std(uncont_normal_finaldec), np.std(uncont_steatotic_finaldec), np.std(uncont_cirrhotic_finaldec)])
 
plt.xlabel('Tissue condition', fontsize = font_size)
plt.ylabel('log10 decorrelation',fontsize = font_size)
xticks = [conditions[0]+', N=' + str(uncont_sample_values[0]),
         conditions[1]+', N=' + str(uncont_sample_values[1]),
         conditions[2]+', N=' + str(uncont_sample_values[2])]
plt.xticks([r for r in range(len(xticks))],xticks)
# plt.title('Final avg log10 decorrelation in ROI, not controlled')
plt.show()
```

```python
# Controlled trials
thresholds = [-2.8, -2.6, -2.5, -2.4, -2.3]

thresholda_normal = [-2.6, -2.8, -2.8, -2.8, -2.8, -2.8]
cont_normal_finaldec = [-3.23, -2.763, -2.793, -2.536, -2.796, -2.965]

thresholda_steatotic = [-2.8, -2.8, -2.8, -2.6, -2.8, -2.8, -2.6]
cont_steatotic_finaldec = [-2.76, -2.757, -2.69, -2.458, -2.752, -2.545, -2.402]

thresholda_cirrhotic = [-2.5, -2.3, -2.3, -2.4]
cont_cirrhotic_finaldec = [-2.48, -2.193, -2.353, -2.875]

fig = plt.figure(figsize = (7, 10))
plt.axes(aspect='equal')
label1 = 'Normal, N=' + str(cont_sample_values[0])
plt.scatter(thresholda_normal, cont_normal_finaldec, c='g', marker='.', s=700, alpha=0.7, label=label1)

label2 = 'Steatotic, N=' + str(cont_sample_values[1])
plt.scatter(thresholda_steatotic, cont_steatotic_finaldec, c='y', marker='+', s=700, alpha=0.7, label=label2)

label3 = 'Cirrhotic, N=' + str(cont_sample_values[2])
plt.scatter(thresholda_cirrhotic, cont_cirrhotic_finaldec, c='r', marker='x', s=700, alpha=0.7, label=label3)

plt.plot(np.linspace(-2.8,-2.3,10), np.linspace(-2.8,-2.3,10), 'k:')
plt.xlabel('Control threshold', fontsize = font_size)
plt.ylabel('Avg log10 decorrelation in ROI', fontsize = font_size)
plt.xticks(thresholds)
# plt.title('Final avg log10 decorrelation in ROI, controlled')
plt.legend(loc='upper left', markerscale=0.5, fontsize = font_size)
plt.show()
```
