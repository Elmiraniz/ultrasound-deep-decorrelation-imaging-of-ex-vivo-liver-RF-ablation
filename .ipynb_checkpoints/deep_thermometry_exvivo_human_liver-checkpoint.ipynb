```python
import tensorflow as tf
from tensorflow import keras
from keras.utils.vis_utils import plot_model
from sklearn.model_selection import train_test_split
import scipy.io as sio
import os
import numpy as np
import matplotlib.pyplot as plt
```

```python
cpath = os.getcwd()
path = '/mat files/instdec_instIBS_temp_4locs_alltrials_except_0528exp3_0608exp1_0708exp1to4.mat'
all_data = sio.loadmat(cpath + path)
all_data.keys()
```

```python
dec_4points = all_data['inst_decorr_4locs']
IBS_4points = all_data['inst_ibs_4locs']
temperature_4points = all_data['temperature_4locs']
```

```python
plt.figure(figsize=(18,10))
plt.subplot(2,3,1)
plt.hist(dec_4points.ravel());
plt.title('Dec')

plt.subplot(2,3,2)
plt.hist(IBS_4points.ravel());
plt.title('IBS')

plt.subplot(2,3,3)
plt.hist(temperature_4points.ravel());
plt.title('Temperature')

plt.subplot(2,3,4)
plt.hist(np.log10(dec_4points.ravel()), range=[-10,2]);
plt.title('log10 dec')

plt.subplot(2,3,5)
plt.hist(np.log10(IBS_4points.ravel()), range=[-10,6]);
plt.title('log10 IBS');
```

### Clean and normalize the data

```python
plt.figure(figsize=(10,3))
plt.subplot(1,3,1)
plt.plot(dec_4points.ravel())
plt.title('Dec');
plt.subplot(1,3,2)
plt.plot(IBS_4points.ravel())
plt.title('IBS');
plt.subplot(1,3,3)
plt.plot(temperature_4points.ravel())
plt.title('Temperature');
```

```python
dec_4points.shape
```

```python
dec_4points = dec_4points.reshape([-1,4])
IBS_4points = IBS_4points.reshape([-1,4])
temperature_4points = temperature_4points.reshape([-1,4])
temperature_4points.shape
```

# Remove nans

```python
dec_4points = dec_4points[~np.isnan(temperature_4points)].reshape([-1,4])
IBS_4points = IBS_4points[~np.isnan(temperature_4points)].reshape([-1,4])
temperature_4points = temperature_4points[~np.isnan(temperature_4points)].reshape([-1,4])

dec_log10 = np.log10(dec_4points)
dec_log10 = np.nan_to_num(dec_log10,posinf=0, neginf=-10)
dec_log10_absmax = abs(dec_log10).max()
dec_log10 = dec_log10 / dec_log10_absmax
dec_log10_mean = np.mean(dec_log10)
dec_log10 = dec_log10 - dec_log10_mean


IBS_log10 = np.log10(IBS_4points)
IBS_log10 = np.nan_to_num(IBS_log10,posinf=0, neginf=-10)
IBS_log10_absmax =  abs(IBS_log10).max()
IBS_log10 = IBS_log10 / IBS_log10_absmax
IBS_log10_mean = np.mean(IBS_log10)
IBS_log10 = IBS_log10 - IBS_log10_mean

temperature_4points_max = temperature_4points.max()
temperature_norm = temperature_4points / temperature_4points_max
temperature_norm_mean = temperature_norm.mean()
temperature_norm = temperature_norm - temperature_norm_mean
temperature_norm.min()
```

```python
plt.figure(figsize=(18,10))
plt.subplot(1,3,1)
plt.hist(dec_log10);
plt.title('Dec normalized');

plt.subplot(1,3,2)
plt.hist(IBS_log10);
plt.title('IBS normalized');

plt.subplot(1,3,3)
plt.hist(temperature_norm);
plt.title('Normalized temperature');
```

```python
dec_IBS = np.concatenate((dec_log10, IBS_log10), axis=1)
dec_IBS.shape
```

```python
X_train_inst, X_test_inst, Y_train_inst, Y_test_inst = train_test_split(dec_IBS, temperature_norm, test_size=0.33, random_state=42)

Y_train_inst.shape
```

### Neural network architecture for instantaneous combined and globally norm. decorr.

```python
my_model = keras.Sequential([
    keras.layers.InputLayer(input_shape=(8,)),
    keras.layers.Dense(units=64, activation='relu'),
    keras.layers.Dense(units=128, activation='relu'),
    keras.layers.Dense(units=64, activation='relu'),
    keras.layers.Dense(units=4)
])

print(my_model.summary())
plot_model(my_model, show_dtype=True, show_layer_activations=True)
```

```python
my_model.compile(optimizer='adam', 
              loss=tf.losses.MeanSquaredError(),
              metrics=[tf.keras.metrics.RootMeanSquaredError()])

history = my_model.fit(
    X_train_inst, 
    Y_train_inst,
    batch_size=50,
    epochs=40, 
    verbose=2,
    validation_split=0.15,
    shuffle=False)
```

```python
import pandas as pd
pd.DataFrame(history.history).plot(figsize=(8,5))
plt.show()
```

```python
y_predictions_inst = my_model.predict(X_test_inst)

Y_test_inst_real = (Y_test_inst + temperature_norm_mean) * temperature_4points_max
y_predictions_inst_real = (y_predictions_inst + temperature_norm_mean) * temperature_4points_max
X_test_inst_dec_real = (X_test_inst[:,0:4] + dec_log10_mean) * dec_log10_absmax
X_test_inst_IBS_real = (X_test_inst[:,4:] + IBS_log10_mean) * IBS_log10_absmax
```
